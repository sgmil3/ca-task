```python
def sqler(q):
    return sqldf(q, globals())
```

```python
concatable = []

for grp, gdf in deliveries.groupby('match_id'):
    print(grp, gdf.shape)
    # if grp != 5460125:
    #     continue
    print(
        q := sqler(
            """
                SELECT 
                    gdf.match_id,
                    gdf.team_innings,
                    team_batting_id,
                    teams_bat.team_name AS batting_team,
                    team_bowling_id,
                    teams_bowl.team_name AS bowling_team,
                    gdf.*,
                    matches.*,
                    (
                        SELECT description
                        FROM lookups
                        WHERE lookups.lookup_type_id = 3 AND lookups.id = matches.match_length_id
                    ) AS match_type,
                    CONCAT(
                        teams_a.team_name, ' vs ', teams_b.team_name
                    ) AS match_name
                FROM 
                    gdf LEFT JOIN matches ON gdf.match_id = matches.match_id
                    LEFT JOIN teams AS teams_bowl ON gdf.team_bowling_id = teams_bowl.team_id
                    LEFT JOIN teams AS teams_bat ON gdf.team_batting_id = teams_bat.team_id
                    LEFT JOIN players AS players_bat ON striker_id = players_bat.player_id
                    LEFT JOIN players AS players_bowl ON bowler_id = players_bowl.player_id
                    LEFT JOIN teams AS teams_a ON teams_a.team_id = matches.team_a_id
                    LEFT JOIN teams AS teams_b ON teams_b.team_id = matches.team_b_id
                ORDER BY gdf.team_innings, team_batting_id, gdf.over, gdf.ball_in_over"""
        )
    )
    q['extras'] = q.apply(lambda x: x[['bye_runs', 'legbye_runs' ,'noball_runs', 'penalty_runs', 'wide_runs']].sum(), axis = 1)
    q['cumulative_ball_number'] = q.apply(lambda x: 6 * (x['over'] - 1) + x['fair_ball_in_over'], axis = 1)
    q['dot_ball'] = q.apply(lambda x: 0 if x['extras'] > 0 or x['bat_score'] > 0 else 1, axis = 1)
    
    concatable.append(q)
```

```python
sqler(
    """
    SELECT 
        d.* ,
        m.*,
        (
            SELECT description
            FROM lookups
            WHERE lookups.lookup_type_id = 3 AND lookups.id = m.match_length_id
        ) AS match_type,
        CONCAT(
            t_a.team_name, ' vs ', t_b.team_name
        ) AS match_name,
        t_a.team_name AS team_a_name,
        t_b.team_name AS team_b_name,
        t_bat.team_name AS team_batting_name,
        t_bowl.team_name AS team_bowling_name,
        l_out.description AS how_out_description
    FROM deliveries d
    LEFT JOIN matches m ON d.match_id = m.match_id
    LEFT JOIN teams t_a ON t_a.team_id = m.team_a_id
    LEFT JOIN teams t_b ON t_b.team_id = m.team_b_id
    LEFT JOIN players p_bat ON d.striker_id = p_bat.player_id
    LEFT JOIN players p_bowl ON d.bowler_id = p_bowl.player_id
    LEFT JOIN teams t_bowl ON d.team_bowling_id = t_bowl.team_id
    LEFT JOIN teams t_bat ON d.team_batting_id = t_bat.team_id
    LEFT JOIN lookups l_out ON l_out.id = d.how_out_id AND l_out.lookup_type_id = 2806
    """
).to_csv(DATA_DIR / "gdf.csv", index=False)
```
